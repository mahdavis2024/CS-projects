```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime as dt
```

```python
cs_excel_path = 'CS timeTable(5March2025).xlsx'
CSDF = pd.read_excel(cs_excel_path)

CSDF
```

```python
CSDF.isna().sum()
```

```python
CSDF = CSDF.dropna(axis= 1, how= 'all')
CSDF.shape
```

```python
CSDF.head(3)
```

```python
CSDF.columns
```

# What about empty row?

```python
CSDF.isna().sum()
```

```python
CSDF.isna().sum(axis=1)
```

```python
CSDF.tail()
```

```python
almost_empty_rows = CSDF.isna().sum(axis=1) > 8
almost_empty_rows
```

```python
almost_empty_rows.sum()
```

```python
CSDF = CSDF[~almost_empty_rows]
CSDF.shape
```

```python
CSDF.isna().sum() 
```

# The data is ready!

```python
CSDF.describe()
```

```python
CS_grouped = CSDF.groupby('Step')
CS_grouped.describe()
```

```python
CS_grouped['Duration'].sum()
```

# It's time to plot!

```python
plt.figure(figsize=(4,4))
sns.histplot(x='Step', data=CSDF, color= 'pink') 
plt.title('Count of Entries per Day');
```

```python
plt.figure(figsize=(16,5))
sns.lineplot(data = CSDF);
```

```python
plt.figure(figsize=(16,4))
sns.countplot(x='Date', data=CSDF)
plt.title('Count of Tasks per Day')
plt.gca().xaxis.set_major_locator(plt.MaxNLocator(nbins=20))
plt.xticks(rotation=45);
```

```python
plt.figure(figsize=(16,5))
CSDF['Duration_minutes'] = CSDF['Duration'].apply(lambda x: x.total_seconds() / 60) 
sns.lineplot(x='Date', y='Duration_minutes', data=CSDF)
plt.legend(['mean value', 'other values'])   # Confidence interval
plt.title('Duration of Tasks or Studies Over Time');
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x='Step', y='Duration_minutes', data=CSDF, color= 'yellow') 
plt.title('Distribution of Task Durations');
```

```python
plt.figure(figsize=(5,3))
sns.boxplot(x='Step', y='Duration_minutes', data=CSDF, color= 'aqua', whis=3)  # Increase the whisker range 
plt.title('Distribution of Task Durations');
```

```python
plt.figure(figsize=(5,3))
sns.violinplot(data=CSDF, x='Step')
plt.title('Distribution of Task Durations');
```

```python
CSDF['time_hours'] = CSDF['Start Time'].apply(lambda x: x.hour + x.minute / 60)
```

```python
plt.figure(figsize=(16, 5))
plt.scatter(CSDF['Date'], CSDF['time_hours']);
```

```python
plt.figure(figsize=(16, 5))
sns.scatterplot(x= CSDF['Date'], y= CSDF['time_hours'], hue= CSDF['Step'], palette= "tab10") #husl

plt.title('Scatter Plot of Date vs Time')
plt.xlabel('Date')
plt.ylabel('Time (hours of the day)')
plt.yticks(range(0, 25))  # Set y-ticks from 0 to 24 hours
plt.grid()
```
